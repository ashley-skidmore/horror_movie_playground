## Cleaning Data

```python
'''I was expecting to see 'release_date' in the stats above.
When I checked its type in the previous cell output, I saw it's an object.
Let's change its type to datetime.'''

horror_movies['release_date'] = pd.to_datetime(horror_movies['release_date'])

#I also just need the year.
horror_movies['year'] = horror_movies['release_date'].dt.year

horror_movies.dtypes
```

```python
#I can check to see what date ranges this data covers.

horror_movies.describe()
```

```python
#I want to look at a few columns specifically, so let's first check to see if those columns have null values.

horror_movies[['title', 'year', 'vote_count', 'vote_average']].isnull().sum()
```

```python
#Even though there aren't nulls in these columns, the stats show me that there are zeros in 'vote_count' and 'vote_average' that should be removed.
horror_filtered = horror_movies[~((horror_movies['vote_count'] == 0) | (horror_movies['vote_average'] == 0))]

'''
Each row is True if at least one has a zero.
The tilde negates this, keeping rows where neither column has a 0.
This filtered version of horror_movies is now stored in horror_filtered.
'''

#Let's check.
horror_filtered.describe()
```

## Plotting Data

### Horror Movie Ratings by Year
Note: This is the data 'as is' regardless of vote count or the number of movies in a given year.

```python
#Grouping the average movie rating by year.
year_ratings = horror_filtered.groupby('year')['vote_average'].mean()

year_ratings.plot(kind='bar', figsize=(15, 6)) #figsize will change the graph size for legibility

#Getting fancy with labels
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Horror Movie Ratings by Year')
plt.show()

```

### Horror Movie Ratings by Decade
Note: This is the data 'as is' regardless of vote count or the number of movies in a given year.

```python
horror_filtered['decade'] = (horror_filtered['year'] // 10) * 10 #Floor division (//) converts each year to a decade (e.g., 1991 to 199) by dividing and rounding down. Multiply by 10 to bring it back to a year (1990). 

decade_ratings = horror_filtered.groupby('decade')['vote_average'].mean() #Calculates the mean for 'vote_average' sorted by each decade

decade_ratings.plot(kind='bar')

#Labels
plt.xlabel('Decade')
plt.ylabel('Average Rating')
plt.title('Average Horror Movie Ratings by Decade')

plt.show()

```

## Horror Movie Ratings Filtered by Number of Votes
### Averages are calculated by ignoring ratings with fewer than a specified number of votes.

```python
#First, I need to see the distribution of votes so I can decide what a reasonable threshold should be.

horror_filtered.plot(kind='scatter', x='year', y='vote_count', title='ratings vote counts')
```

```python
#I can't see what the smaller values are, which may cause me to accidentally eliminate certain years. Let's change the y axis limits.

horror_filtered.plot(kind='scatter', x='year', y='vote_count', title='ratings vote counts')
plt.ylim(0, 1000)
plt.show()
```

```python
#Let's look more closely at the area between 0 and 200. That's probably where I'll have to select a vote threshold to accommodate all years.

horror_filtered.plot(kind='scatter', x='year', y='vote_count', title='ratings vote counts')
plt.ylim(0, 200)
plt.show()
```

```python
#Looks like I will set my threshold at 20 to be inclusive of all years.

filtered_votes = horror_filtered[horror_filtered['vote_count'] >= 20]

#And now we do the bar plot all over again using the new DataFrame filtered_votes

filtered_year_ratings = filtered_votes.groupby('year')['vote_average'].mean()

filtered_year_ratings.plot(kind='bar', figsize=(15, 6))

plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Horror Movie Ratings by Year(>= 20 Votes)')
plt.show()





```

```python
#This is a little inconvenient because I want to compare this to the unfiltered bar graph, but it's high up in the code. Let's put them side by side.

#Creating a figure with two subplots:
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

#Plotting year_ratings:
axes[0].bar(year_ratings.index, year_ratings.values)
axes[0].set_title('Average Rating by Year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Average Rating')

#Plotting filtered_year_ratings:
axes[1].bar(filtered_year_ratings.index, filtered_year_ratings.values)
axes[1].set_title('Average Rating by Year (>=20 Votes)')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Average Rating')

#Avoiding overlap by altering the layout
plt.tight_layout()

plt.show()

#Looking at this, they're virtually the same. That's probably because 20 is such a low threshold for votes.
```

```python
#Let's change the threshold to something higher with the understanding that certain years won't meet the criteria.

#Redefining filtered_votes to only include ratings where there were at least a thousand votes.
filtered_votes = horror_filtered[horror_filtered['vote_count'] >= 1000]

filtered_year_ratings = filtered_votes.groupby('year')['vote_average'].mean()

#Pasting the code in the above cell to display the two graphs side by side, this time with the new vote threshold.
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

#Plotting year_ratings:
axes[0].bar(year_ratings.index, year_ratings.values)
axes[0].set_title('Average Rating by Year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Average Rating')

#Plotting filtered_year_ratings:
axes[1].bar(filtered_year_ratings.index, filtered_year_ratings.values)
axes[1].set_title('Average Rating by Year (>=1000 Votes)')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Average Rating')

#Avoiding overlap by altering the layout
plt.tight_layout()

plt.show()

#That's better. Looks like the mid-90s has a year where not many people watched and reviewed the movies. Also, 1960 is a clear high point when the votes below 1000 are filtered out. The 70s aren't too shabby either.
```
