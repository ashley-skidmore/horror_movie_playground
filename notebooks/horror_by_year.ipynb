## Cleaning Data

```python
'''I was expecting to see 'release_date' in the stats above.
When I checked its type in the previous cell output, I saw it's an object.
Let's change its type to datetime.'''

horror_movies['release_date'] = pd.to_datetime(horror_movies['release_date'])

#I also just need the year.
horror_movies['year'] = horror_movies['release_date'].dt.year

horror_movies.dtypes
```

```python
#I want to look at a few columns specifically, so let's first check to see if those columns have null values.

horror_movies[['title', 'year', 'vote_count', 'vote_average']].isnull().sum()
```

```python
#Even though there aren't nulls in these columns, the stats show me that there are zeros in 'vote_count' and 'vote_average' that should be removed.
horror_filtered = horror_movies[~((horror_movies['vote_count'] == 0) | (horror_movies['vote_average'] == 0))].copy() 
#I had to add .copy() to the end to prevent a message later on telling me it's unclear if horror_filtered is an independent DataFrame or a temporary slice/view.

'''
Each row is True if at least one has a zero.
The tilde negates this, keeping rows where neither column has a 0.
This filtered version of horror_movies is now stored in horror_filtered.
'''

#Let's check.
horror_filtered.describe()
```

## Standard Error, Confidence Intervals, and Weighted Values

### Standard Error
How reliable is the average horror movie rating each year?
- Some years have few movies
- Some movie ratings have few votes  
This is an example of **sampling variability**--the idea that small samples can give misleading averages.

**Standard Error**: How much the average might vary due to sample randomness
For example, a smaller sample size may vary more, so the standard error is higher(a.k.a., less confidence).

*std deviation of ratings / sqrt(number of movies in that year)*

### Confidence Intervals

**Confidence Interval**: The range likely containing the true average
- If the standard error is big, the confidence interval is wide (more uncertainty).
- If the standard error is small, the confidence interval is tight (more precise).

These intervals provide **error bars** on the plot displaying the uncertainty of each category.

*lower = mean - 1.96 * standard_error  
upper = mean + 1.96 * standard_error*

```python
'''
filtered_year_ratings has the vote_average_mean grouped by year, 
but it only contains one value per movie.
I need to be able to calulate standard deviation and number of movies as well.
That's why I had to go back to regroup everything without limiting it to the mean.
'''
grouped = horror_filtered.groupby('year')['vote_average']

#Calculating the standard error based on number of movies in a year
se = grouped.std() / np.sqrt(grouped.count())

#Calculating the mean for each year
mean_ratings = grouped.mean()

#Calculating the 95% confidence interval using standard error
ci_lower = mean_ratings - 1.96 * se
ci_higher = mean_ratings + 1.96 * se
```

```python
# Plot the mean with confidence intervals as error bars
plt.errorbar(mean_ratings.index, mean_ratings, yerr=1.96 * se, fmt='-o', ecolor='gray', capsize=3) 
'''
yerr is y axis error bar and defines the error bars size
fmt is format style, - makes it a solid line, and o circles each data point
ecolor defines the color of the error bars
capsize defines the length of the horizontal lines at the tops and bottoms of the error bars
'''

# Labels and title
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Horror Movie Ratings by Year with Confidence Intervals')

plt.show()
```

```python
#The plot above seemed cramped, so I'm going to widen it.

plt.figure(figsize=(20,6))

plt.errorbar(mean_ratings.index, mean_ratings, yerr=1.96 * se, fmt='-o', ecolor='gray', capsize=3) 

plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Horror Movie Ratings by Year with Confidence Intervals')

plt.show()
```

### Adding a Weighted Value

The plots above account for the number of movies each year.  
What if I also want to take the number of votes per movie into consideration?  

I need to:
1. Use a weighted average for the column vote_average instead of a mean.
2. Calculated a weighted standard error.  

<u>**This ended up being way more complex than I could figure, so just for enlightenment I got the following result from ChatGPT.  
The code is generated, but the documentation is my own as I picked through each line for understanding.**<u>

```python
#Defining a function to handle the weighted mean
def weighted_mean(x): #x is a subset of the grouped DataFrame representing the rows for each year
    return np.average(x['vote_average'], weights=x['vote_count']) #For each movie, we're multiplying the 'vote_average' by the 'vote_count' and dividing by the total votes for the year

#Group by year and apply the function to the new variable
grouped = horror_filtered.groupby('year')
weighted_means = grouped.apply(weighted_mean)

#Define a function to calculate weighted standard error
def weighted_se(x):
    #assigns 'weights' the value of votes per year
    weights = x['vote_count']
    #calculating a weighted average of x['vote_average'] using its value in 'weights'
    average = np.average(x['vote_average'], weights=weights) 
    #calculating the difference between each rating and its weighted rating, squaring to eliminate negatives and emphasize large differences, then calculating an average 
    variance = np.average((x['vote_average'] - average) ** 2, weights=weights) 
    #FINALLY we calculate the weighted standard error by multiplying variance by the number of movies and finding the square root of that result
    return np.sqrt(variance / len(x))

#Apply the function to calculate the weighted standard error for grouped
weighted_se = grouped.apply(weighted_se)

#Calculate confidence interval
ci_lower = weighted_means - 1.96 * weighted_se
ci_upper = weighted_means + 1.96 * weighted_se
```

```python
plt.figure(figsize=(20,6))
plt.errorbar(weighted_means.index, weighted_means, yerr=1.96 * weighted_se, fmt='-o', ecolor='gray', capsize=3)

plt.xlabel('Year')
plt.ylabel('Weighted Average Rating')
plt.title('Weighted Average Horror Movie Ratings by Year with 95% Confidence Intervals')

plt.show()
```
